# Feature Engineering- Flight Price data Set

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('flight_price.xlsx')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop('Route',axis=1,inplace=True)
df.head()
```

```python
#Handling Date- Splitting the date_of_journey to Date_j, Month_j,Year_j
df['Date']=(df['Date_of_Journey'].str.split('/').str[0]).astype(int)
df['Month']=(df['Date_of_Journey'].str.split('/').str[1]).astype(int)
df['Year']=(df['Date_of_Journey'].str.split('/').str[2]).astype(int)
```

```python
df.head()
```

```python
#Droping the Date_of_Journey
df.drop('Date_of_Journey',axis=1,inplace=True)
df.head()
```

```python
#Handling the Arrival time And Dep_Time
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df['Arrival_hour']=(df['Arrival_Time'].str.split(':').str[0]).astype(int)
df['Arrival_min']=(df['Arrival_Time'].str.split(':').str[1]).astype(int)
df.head()
```

```python
#Droping the Arrival_Time
df.drop('Arrival_Time',axis=1,inplace=True)
df.head()
```

```python
df['Dep_Time']=df['Dep_Time'].apply(lambda x:x.split(' ')[0])
df['Dep_hour']=(df['Dep_Time'].str.split(':').str[0]).astype(int)
df['Dep_min']=(df['Dep_Time'].str.split(':').str[1]).astype(int)
df.head()
```

```python
#Droping the Dep_Time
df.drop('Dep_Time',axis=1,inplace=True)
df.head()
```

```python
df.info()
```

```python
df['Duration_hour']=df['Duration'].str.split('h').str[0]
df['Duration_min']=df['Duration'].str.split(' ').str[1].str.replace('m', '')
df.head()
```

```python
df['Duration_min'].isnull().sum()
```

```python
df['Duration_min']=df['Duration_min'].fillna('0')
```

```python
df['Duration_hour']=df['Duration_min'].fillna('0')
```

```python
df['Duration_min'].isnull().sum()
```

```python
df['Duration_min']=df['Duration_min'].astype(int)
df['Duration_hour']=df['Duration_hour'].astype(int)
```

```python
df.info()
```

```python
#Droping the Duration
df.drop('Duration',axis=1,inplace=True)
df.head()
```

```python
df.info()
```

```python
#Handling    Total_Stops  Feature
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '1 stop':1,'2 stops':2, '3 stops':3, '4 stops':4,np.nan:1})
```

```python
df.head()
```
